```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('Chinook_Sqlite.sqlite')
```

```python
c = conn.cursor()
```

```python
def prettify(c):
    df = pd.DataFrame(c.fetchall())
    df.columns = [i[0] for i in c.description]
    return df
```

```python
c.execute('''SELECT name FROM sqlite_master
        WHERE type ='table' ''')
prettify(c)
```

```python
c.execute("""
    SELECT *
    FROM Invoice
    """)
prettify(c).head()
```

```python
c.execute("""
    SELECT *
    FROM Customer
    """)
prettify(c).head()
```

```python
c.execute("""
    SELECT CustomerId, AVG(Total) as AvgPurchase, FirstName, LastName
    FROM Invoice
    JOIN Customer USING(CustomerId)
    GROUP BY CustomerId
    ORDER BY AvgPurchase DESC
    LIMIT 3
    """)
prettify(c)
```
