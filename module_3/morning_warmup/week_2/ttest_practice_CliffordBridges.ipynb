```python
import pandas as pd
import scipy.stats as stats
```

```python
car_df = pd.read_csv('auto-mpg.csv')
```

```python
car_df.head()
```

```python
cars_1970_1976 = car_df.loc[(car_df['model year']<=76) & (car_df['model year']>=70)]
```

```python
cars_1977_1982 = car_df.loc[(car_df['model year']<=82) & (car_df['model year']>=77)]
```

```python
stats.t.ppf(0.95, len(cars_1977_1982.cylinders)-1)
```

```python
# stats.ttest_1samp
stats.ttest_ind(cars_1970_1976.cylinders, cars_1977_1982.cylinders, equal_var=False)
# stats.ttest_ind_from_stats
# stats.ttest_rel
```
