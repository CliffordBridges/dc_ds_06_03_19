```python
bananasaver = pd.read_html('http://bananasaver.com/OrderOnline.htm')
```

```python
len(bananasaver)
```

```python
dfs = {}
for df in range(len(bananasaver)):
    dfs[df] = pd.DataFrame(bananasaver[df])
```

```python
dfs[2].head()
```

```python
bananasaver_df = dfs[2]
```

```python
bananasaver_df.head()
```

```python
bananasaver_df = bananasaver_df.drop(0, axis=1)
bananasaver_df = bananasaver_df.rename(columns=bananasaver_df.iloc[0])
bananasaver_df.head()
```

```python
bananasaver_df = bananasaver_df.drop(0)
```

```python
bananasaver_df.head()
```

```python
bananasaver_df['ShippingCost'] = bananasaver_df.iloc[:,1]
```

```python
bananasaver_df.ShippingCost = bananasaver_df.ShippingCost.map(lambda x: float(x[1:]))
```

```python
bananasaver_df.ShippingCost.describe()
```

```python
bananasaver_df.Quantity = bananasaver_df.Quantity.map(lambda x: float(x[-len('4.25 ea.)'):-len('4.25 ea.)')+4]))
```

```python
bananasaver_df.head()
```

```python
bananasaver_df.describe()
```

```python
# bananasaver_df.rename({'Quantity':'PricePerSaver'}, inplace=True)
bananasaver_df['PricePerSaver'] = bananasaver_df.iloc[:,1]

bananasaver_df.index.name = 'Quantity'
```

```python
bananasaver_df.head()
```

```python
bananasaver_df['TotalCost'] = (bananasaver_df.Quantity)*bananasaver_df.PricePerSaver + bananasaver_df.ShippingCost
```

```python
plt.bar(bananasaver_df.index, bananasaver_df.TotalCost)
plt.show()
```

```python
bananasaver_df['AverageCost'] = bananasaver_df.TotalCost/bananasaver_df.index
```

```python
plt.bar(bananasaver_df.index, bananasaver_df.AverageCost)
plt.show()
```
