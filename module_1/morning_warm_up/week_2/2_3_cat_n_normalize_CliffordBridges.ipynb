```python
# Normalize data between 0 and 1
def normalize_df_column(df, column_name):
    center = df[column_name].mean()
    stddev = df[column_name].std()
    minimum = df[column_name].min()
    stretch = df[column_name].max() - df[column_name].min()
    print(f'This data is centered around {center}, with a standard deviation of of {stddev}.')
    df[column_name+' Normalized'] = df[column_name].map(lambda x: (x-minimum)/stretch)
    return df
```

```python
normalize_df_column(CO2_df, 'Interpolated Average').head()
```
